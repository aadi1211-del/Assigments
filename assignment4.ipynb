# Task 1: Write Sales Records to a File

```python
# Task 1: Write Sales Records to a File

# Create a list of sales amounts
sales = [1200, 450, 980, 1500, 3000]

# Open file in write mode
file = open('sales_data.txt', 'w')

# Write each sale on a new line in the file
for sale in sales:
    file.write(str(sale) + '\n')

# Close the file
file.close()

# Reopen and print the contents
print("Task 1: Write Sales Records to a File")
print("=" * 50)

file = open('sales_data.txt', 'r')
contents = file.read()
file.close()

print("File contents:")
print(contents)
print("\nFile 'sales_data.txt' has been created successfully.")
```

# Task 2: Read File in Different Ways

```python
# Task 2: Read File in Different Ways

print("Task 2: Read File in Different Ways")
print("=" * 50)

# First, create the sales_data.txt file
sales = [1200, 450, 980, 1500, 3000]
file = open('sales_data.txt', 'w')
for sale in sales:
    file.write(str(sale) + '\n')
file.close()

# Method 1: Read entire file using read()
print("\nMethod 1: Using read() - Entire file as one string:")
print("-" * 50)
file = open('sales_data.txt', 'r')
entire_content = file.read()
file.close()
print(entire_content)

# Method 2: Read first line using readline()
print("Method 2: Using readline() - First line only:")
print("-" * 50)
file = open('sales_data.txt', 'r')
first_line = file.readline()
file.close()
print(f"First line: {first_line.strip()}")

# Method 3: Read all lines using readlines() and convert to integers
print("\nMethod 3: Using readlines() - All lines as list of integers:")
print("-" * 50)
file = open('sales_data.txt', 'r')
all_lines = file.readlines()
file.close()

# Convert to integers and remove newline characters
sales_integers = [int(line.strip()) for line in all_lines]
print(f"Sales as integers: {sales_integers}")
print(f"List of integers: {sales_integers}")
```

# Task 3: Append New Sales to File

```python
# Task 3: Append New Sales to File

print("Task 3: Append New Sales to File")
print("=" * 50)

# First, create the initial sales_data.txt file
initial_sales = [1200, 450, 980, 1500, 3000]
file = open('sales_data.txt', 'w')
for sale in initial_sales:
    file.write(str(sale) + '\n')
file.close()

print("Initial file created with sales:")
print(initial_sales)

# New sales to append
new_sales = [5000, 2500, 1700]

# Append new sales to the file
print(f"\nAppending new sales: {new_sales}")
file = open('sales_data.txt', 'a')
for sale in new_sales:
    file.write(str(sale) + '\n')
file.close()

print("New sales appended successfully.")

# Reopen and print all contents
print("\nAfter appending, file contents:")
print("-" * 50)
file = open('sales_data.txt', 'r')
all_lines = file.readlines()
file.close()

# Print all lines and count them
print(f"All sales in file:")
for line in all_lines:
    print(line.strip())

print(f"\nTotal number of lines after appending: {len(all_lines)}")
```

# Task 4: Generate Summary Report from File

```python
# Task 4: Generate Summary Report from File

print("Task 4: Generate Summary Report from File")
print("=" * 50)

# First, create the sales_data.txt file
sales = [1200, 450, 980, 1500, 3000]
file = open('sales_data.txt', 'w')
for sale in sales:
    file.write(str(sale) + '\n')
file.close()

# Read all sales values from file
file = open('sales_data.txt', 'r')
all_lines = file.readlines()
file.close()

# Convert to integers
sales_integers = [int(line.strip()) for line in all_lines]

# Calculate statistics
total_sales = sum(sales_integers)
highest_sale = max(sales_integers)
lowest_sale = min(sales_integers)
average_sale = total_sales / len(sales_integers)

# Print the summary report
print("\nSales Summary Report")
print("-" * 50)
print(f"Total Sales: {total_sales}")
print(f"Highest Sale: {highest_sale}")
print(f"Lowest Sale: {lowest_sale}")
print(f"Average Sale: {average_sale:.2f}")
print(f"Number of Sales: {len(sales_integers)}")
```

# Task 5: Create Product Info File (User Input)

```python
# Task 5: Create Product Info File (User Input)

print("Task 5: Create Product Info File (User Input)")
print("=" * 50)

# Ask user for 3 product names and their prices
products = {}

print("\nEnter 3 product names and their prices:")
print("-" * 50)

for i in range(3):
    product_name = input(f"Enter product {i+1} name: ").strip()

    try:
        product_price = float(input(f"Enter price for {product_name}: "))
        if product_price < 0:
            print("Error: Price cannot be negative.")
            i -= 1
            continue
        products[product_name] = product_price
    except ValueError:
        print("Error: Invalid price. Please enter a valid number.")
        i -= 1
        continue

# Write products to file
file = open('products.txt', 'w')
for product_name, product_price in products.items():
    file.write(f"{product_name} | {product_price}\n")
file.close()

print("\nProducts written to 'products.txt'")

# Read the file and print each line with proper formatting
print("\nReading products from file:")
print("-" * 50)
file = open('products.txt', 'r')
lines = file.readlines()
file.close()

print(f"{'Product Name':<20} {'Price':<10}")
print("-" * 30)
for line in lines:
    parts = line.strip().split(' | ')
    if len(parts) == 2:
        product_name, price = parts
        print(f"{product_name:<20} {float(price):<10.2f}")
```

# Task 7: Mini Project - Export Discounted Prices

```python
# Task 7: Mini Project - Export Discounted Prices

print("Task 7: Mini Project - Export Discounted Prices")
print("=" * 60)

# Create a dictionary with product prices
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

# Ask user for discount percentage
print("\nEnter discount percentage for products:")
discount_percent = float(input("Enter discount percentage (0-100): "))

# Validate discount
if discount_percent < 0 or discount_percent > 100:
    print("Invalid discount percentage. Using 0%.")
    discount_percent = 0

print(f"\nApplying {discount_percent}% discount to all products...")

# Calculate discounted prices
discounted_data = []

for product, price in prices.items():
    discount_amount = price * (discount_percent / 100)
    discounted_price = price - discount_amount
    discounted_data.append((product, price, discounted_price))

# Write discounted prices to file
file = open('discount_report.txt', 'w')
for product, original_price, discounted_price in discounted_data:
    file.write(f"{product} | {original_price} | {discounted_price}\n")
file.close()

print("\nDiscounted prices written to 'discount_report.txt'")

# Read the file and print to terminal
print("\nReading discount report from file:")
print("-" * 60)

file = open('discount_report.txt', 'r')
lines = file.readlines()
file.close()

# Print formatted report
print(f"{'Product':<15} {'Original Price':<20} {'Discounted Price':<20}")
print("-" * 55)

total_items = 0
total_discounted_price = 0

for line in lines:
    parts = line.strip().split(' | ')
    if len(parts) == 3:
        product, original_price, discounted_price = parts
        original = float(original_price)
        discounted = float(discounted_price)
        print(f"{product:<15} {original:<20.2f} {discounted:<20.2f}")
        total_items += 1
        total_discounted_price += discounted

print("-" * 55)
print(f"\nSummary at bottom of file:")
print(f"Total Items: {total_items}")
print(f"Average Discounted Price: {total_discounted_price / total_items:.2f}")
```
