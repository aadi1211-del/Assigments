# Task 1: Discount Rules (if / elif / else)

```python
order_amount = int(input("Enter order amount: "))

# Ensure input is non-negative
if order_amount < 0:
    print("Error: Order amount cannot be negative.")
    exit()

# Apply discount rules
if order_amount >= 2000:
    discount_rate = 15
elif order_amount >= 1500:
    discount_rate = 10
elif order_amount >= 1000:
    discount_rate = 7
else:
    discount_rate = 0

# Calculate discount and final amount
discount = order_amount * (discount_rate / 100)
final_amount = order_amount - discount

# Print results
print(f"Order Amount: {order_amount}")
print(f"Discount Rate: {discount_rate}%")
print(f"Discount: {discount}")
print(f"Final Amount: {final_amount}")

# Optional: Add tax (fixed 5%) after discount and print subtotal, tax, and final total
tax_rate = 5
subtotal = final_amount
tax = subtotal * (tax_rate / 100)
total = subtotal + tax

print(f"Subtotal (after discount): {subtotal}")
print(f"Tax ({tax_rate}%): {tax}")
print(f"Final Total: {total}")
```

# Task 2: Process Multiple Orders (for loop)

```python
# Task 2: Process Multiple Orders (for loop)

orders = [1200, 2500, 800, 1750, 3000]

print("Order Processing Summary")
print("=" * 60)
print(f"{'Order Amount':<15} {'Discount %':<12} {'Discount':<15} {'Final Amount':<15}")
print("=" * 60)

total_revenue = 0
discounted_orders = 0

for order_amount in orders:
    # Apply discount rules
    if order_amount >= 2000:
        discount_rate = 15
    elif order_amount >= 1500:
        discount_rate = 10
    elif order_amount >= 1000:
        discount_rate = 7
    else:
        discount_rate = 0

    # Calculate discount and final amount
    discount = order_amount * (discount_rate / 100)
    final_amount = order_amount - discount

    # Add to total revenue
    total_revenue += final_amount

    # Count orders with discount
    if discount > 0:
        discounted_orders += 1

    # Print order details
    print(f"{order_amount:<15} {discount_rate:<12} {discount:<15.2f} {final_amount:<15.2f}")

print("=" * 60)
print(f"Total Revenue after Discounts: {total_revenue:.2f}")
print(f"Number of Orders with Discount: {discounted_orders}")
```

# Task 3: User Menu (while loop + break/continue)

```python
# Task 3: User Menu (while loop + break/continue)

orders_list = []
total_revenue = 0

while True:
    print("\n" + "=" * 40)
    print("Menu Options:")
    print("1 — Add order amount to a running list (use input())")
    print("2 — Show all orders and totals after applying discounts")
    print("q — Quit")
    print("=" * 40)

    choice = input("Enter your choice (1, 2, or q): ").strip().lower()

    if choice == 'q':
        print("Exiting the program. Goodbye!")
        break

    elif choice == '1':
        try:
            order_amount = int(input("Enter order amount: "))

            if order_amount < 0:
                print("Error: Order amount cannot be negative.")
                continue

            orders_list.append(order_amount)
            print(f"Order added: {order_amount}")
        except ValueError:
            print("Error: Invalid input. Please enter a valid number.")
            continue

    elif choice == '2':
        if not orders_list:
            print("No orders to display.")
            continue

        print("\n" + "=" * 60)
        print("Order Summary with Discounts Applied")
        print("=" * 60)
        print(f"{'Order Amount':<15} {'Discount %':<12} {'Discount':<15} {'Final Amount':<15}")
        print("=" * 60)

        total_revenue = 0

        for order_amount in orders_list:
            # Apply discount rules
            if order_amount >= 2000:
                discount_rate = 15
            elif order_amount >= 1500:
                discount_rate = 10
            elif order_amount >= 1000:
                discount_rate = 7
            else:
                discount_rate = 0

            # Calculate discount and final amount
            discount = order_amount * (discount_rate / 100)
            final_amount = order_amount - discount

            # Add to total revenue
            total_revenue += final_amount

            # Print order details
            print(f"{order_amount:<15} {discount_rate:<12} {discount:<15.2f} {final_amount:<15.2f}")

        print("=" * 60)
        print(f"Total Revenue after Discounts: {total_revenue:.2f}")

    else:
        print("Invalid input. Please choose 1, 2, or q.")
```

# Task 4: Loop Control with Conditions (break & continue)

```python
# Task 4: Loop Control with Conditions (break & continue)

daily = [200, 0, 400, 50, -1, 300]

total_sales = 0

print("Daily Sales Processing")
print("=" * 40)

for day_sales in daily:
    # If a day's value is -1, treat it as corrupted data and break
    if day_sales == -1:
        print(f"Day Sales: {day_sales} — Corrupted data detected. Stopping.")
        break

    # If a day's value is 0, skip (no sales that day) and continue
    if day_sales == 0:
        print(f"Day Sales: {day_sales} — No sales today. Skipping.")
        continue

    # For valid positive sales, add to total
    if day_sales > 0:
        total_sales += day_sales
        print(f"Day Sales: {day_sales} — Added to total.")

print("=" * 40)
print(f"Final Total Sales: {total_sales}")
```
